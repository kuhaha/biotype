```python
import pickle
import pandas as pd
import biotype.similarity as sim
import biotype.peak_extract as pke

DATA_DIR_PATH = "datasets/export"
META_DIR_PATH = "doc"
PEAK_DIR_PATH = "extracted_peaks"
files =[
    "L185_0_A9_1", 
    "L101_0_A1_1"
]

extracted = []
for i in range(2):
    df = None
    df = pd.read_table(f"{DATA_DIR_PATH}/{files[i]}.txt",sep=" ", header=None,names=['m/z', 'intensity']) 
    x, y = df['m/z'].to_numpy(), df['intensity'].to_numpy()

    with open(f'{PEAK_DIR_PATH}/{files[i]}_peaks.pkl', 'rb') as peak_file:
        peaks = pickle.load(peak_file)
        print(f'{files[i]}, peaks: {len(peaks)}')
        extracted +=  [(x[peaks], y[peaks]) ] 

print(extracted[0][0][:5])
print(extracted[1][0][:5])
```

```python
pk1, pk2 = extracted[0], extracted[1]

# transform a list of <m/z, intensity> to a list of <peak_id, intensity>
# where m/z values close enough are assigned an identical peak_id
npk1, npk2 = pke.align(pk1, pk2)

for m in npk1:
    print (f"{m}: {npk1[m]}")
```
